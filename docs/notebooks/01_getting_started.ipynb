# Capper: Getting started

Build a Pydantic model with Capper types; Polyfactory generates fake data with no extra config.

```python
from pydantic import BaseModel
from capper import Name, Email
from polyfactory.factories.pydantic_factory import ModelFactory


class User(BaseModel):
    name: Name
    email: Email


class UserFactory(ModelFactory[User]):
    pass


user = UserFactory.build()
print("Name:", user.name)
print("Email:", user.email)
```

Run the cell again to get different values. Capper types are subclasses of `str`, so you can use them anywhere you'd use a string.
