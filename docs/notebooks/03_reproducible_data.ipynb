# Reproducible data (seeding)

Use `capper.seed(n)` so the same seed produces the same fake data. Useful for tests and demos.

```python
from capper import seed, Name, Email
from polyfactory.factories.pydantic_factory import ModelFactory
from pydantic import BaseModel


class User(BaseModel):
    name: Name
    email: Email


class UserFactory(ModelFactory[User]):
    pass


seed(42)
user1 = UserFactory.build()
seed(42)
user2 = UserFactory.build()

print("user1:", user1.name, user1.email)
print("user2:", user2.name, user2.email)
print("Same?", user1.name == user2.name and user1.email == user2.email)
```

You can also use `UserFactory.seed_random(42)` before `build()` — it seeds the same shared Faker instance.

```python
UserFactory.seed_random(123)
a = UserFactory.build()
seed(123)
b = UserFactory.build()
print("Same name?", a.name == b.name)
```
